#**1. Data Exploration**#

From the Kaggle data description:

>The data files train.csv and test.csv contain gray-scale images of hand-drawn digits, from zero through nine.
>
> Each image is 28 pixels in height and 28 pixels in width, for a total of 784 pixels in total. Each pixel has a single pixel-value associated with it, indicating the lightness or darkness of that pixel, with higher numbers meaning darker. This pixel-value is an integer between 0 and 255, inclusive.
>
> The training data set, (train.csv), has 785 columns. The first column, called "label", is the digit that was drawn by the user. The rest of the columns contain the pixel-values of the associated image.

First, let's check that we have an even distribution of digits in our training data.

```r
barplot(table(digits.train[, 1]), col = "blue", main = "Count of Digits in Training Set")
```

From the data description, we know what the format of the training data will be (and we'll take Kaggle's word for it). However, to make things easier, we should scale the pixel data to $[0, 1]$. Then we will display a few of images to to see quickly see what we're dealing with. We'll visually the first 12 images of each label.

```r
##Scale the pixel data
pixels.train <- digits.train[,-1]/255
pixels.test <- digits.test/255

##Scaled training set
train.scaled <- cbind('label' = digits.train$label,
                      pixels.train)

##Display a 10x12 matrix of images
par(mfrow=c(10, 12), pty='s', mai=c(0.2, 0, 0, 0.1))

for (lab in 0:9) {
  samp <- train.scaled %>%
    filter(label == lab)
  for (i in 1:12) {
    img <- matrix(as.numeric(samp[i, -1]), 28, 28, byrow = TRUE)
    image(t(img)[,28:1], axes = FALSE, col = grey(seq(1, 0, length = 256)))
    box(lty = 'solid')
  }
}
```

We can see from this sample that there are significant differences between the images of the same digit. Some issues that immediately pop out are:

* Digits have different "slants", particularly visible in the digit *1*
* We see different styles of writing. The digit *7* and *2* are good examples. 
* "Poor" handwritin maybe an issue we may need to address in our classification algorithm. There are a couple digits that are unrecognizable, at least to me.
* Some digits are are more faded than others. Some are "thicker"

While there are some issues, we also notice that the data is generally of good quality. There aren't any misclassifications that we see from this sample. The digits are also written "correctly"; there no digits written backwards for example. Lastly, we see that written digits are generally horizontally centered.

Lets look at the "average" of each image.

```r
par(mfrow=c(2, 5), pty='s', mai=c(0, 0, 0, 0))

for (lab in 0:9) {
  subs <- train.scaled %>%
    filter(label == lab)
  avg <- colMeans(subs)
  img <- matrix(avg[2:length(avg)], 28, 28, byrow = TRUE)
  image(t(img)[,28:1], axes = FALSE, col = grey(seq(1, 0, length = 256)))
  box(lty = 'solid')
}
```

These confirm the the digits are generally centered and not flipped. Most importantly images of the same digit have the same generally features. Lastly, these give us a rough guide we can use to augment the data (how much to rotate, shift, etc).

#A Simple Neural Network#

Since it looks like images of the same digit have similar features, judging from the average of the images, a simple feed-forward neural network should perform pretty well. We'll test this out using the Keras library with the TensorFlow backend. The code here is run in the Kaggle Kernal with **GPU set to *ON***.

Using Keras we'll build a simple sequential model with a single hidden layer. However, the data is not quite ready just yet. We need to perform some preprocessing to get the data into the format used by the Keras functions. First, the classes matrix need to be in a binary format, but the *label* column in our data is a vector in the range 0 to 9. We need to *One-Hot Encode* this.



```r
one_hot_labels <- to_categorical(digits.train$label, 10)
dimnames(one_hot_labels) <- list(NULL, c(0:9))
```

We already scaled the *pixels*, but we also need to convert these into matrix, it's currently a data frame.

```r
pixels.train <- as.matrix(digits.train[,-1]/255)
pixels.test <- as.matrix(digits.test/255)
```

Finally, we're ready to build our Neural Network:

```r
NN <- keras_model_sequential()
NN %>%
    layer_dense(units = 128, activation = "relu", input_shape = c(784)) %>%
    layer_dense(units = 10, activation = "softmax") 
    
NN %>%
    compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = "accuracy")
```

Notice that this is a pretty canonical neural net and possibly quite naive but we're just testing the water at this point. Next, we'll train this model:

```r
tic("Fitting Model")
history <- NN %>% 
    fit(pixels.train, one_hot_labels, 
        epochs = 30, 
        validation_split = 0.1, 
        batch_size = 64, 
        verbose = 2)
toc()
```

With the GPU on, this took less than 2 minutes in the Kaggle Kernel. Lets visually evaluate this model by examining the loss and accuracy curves

```r
plot(history)
```

After about 10 epoch, we can see the performance start to suffer as validation loss is starting to increase and validiation accuracy start to plateau. We can do a lot to optimize this model but that's more than I want to get into for this notebook.
